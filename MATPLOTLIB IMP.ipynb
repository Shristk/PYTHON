```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y,marker='*')
plt.xticks(x,labels=["python","c","c++","java"])
plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,10)


plt.show()
```
